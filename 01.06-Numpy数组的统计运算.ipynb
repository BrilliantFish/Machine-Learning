```python
import numpy as np
```

## 求最值

```python
X = np.arange(1000)
```

```python
X.max()
```

```python
X.min()
```

```python
Y = X.reshape(5, -1)
Y
```

```python
np.min(Y, axis = 0)
```

```python
np.min(Y, axis = 1)
```

## 求总和

```python
np.sum(Y)
```

```python
np.sum(Y, axis = 0)
```

```python
np.sum(Y, axis = 1)
```

## 求均值

```python
np.mean(Y)
```

## 求中位数

```python
np.median(Y)
```

## 求标准差

```python
np.std(Y)
```

## 求方差

```python
np.var(Y)
```

## 最大值与最小值的差

```python
np.ptp(Y)
```

## 百分位数，中位数就是50%位的数

```python
q = np.linspace(0, 100, 5)
q, np.percentile(Y, q)
```

## 累加

```python
np.cumsum(Y, axis=0)
```

```python
np.cumsum(Y, axis=1)
```

## 相邻的差 

```python
np.diff(Y, axis = 0)
```

```python
np.cumsum(Y, axis=1)
```

## 乘积

```python
np.prod(Y)
```

```python
Y, np.prod(Y, axis = 0)
```
