```python
x = np.linspace(-10, 10, 100)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1, color = 'green') # 
plt.plot(x,y2, linestyle = ':') # 

plt.show()
```

```python
plt.plot(x,y1, color = 'green') # 
plt.plot(x,y2, linestyle = ':') # 

# x,y
plt.xlim(-5,5)
plt.ylim(-2,2)

plt.show()
```

```python
plt.plot(x,y1, color = 'green') # 
plt.plot(x,y2, linestyle = ':') # 

# x,y
# plt.xlim(-5,5)
# plt.ylim(-2,2)
# 
plt.axis([-5,5,-2,2])

plt.show()
```

## X,Y & 

```python
plt.plot(x,y1, color = 'green', label='sin(x)') # 
plt.plot(x,y2, linestyle = ':', label='cos(x)') # 

plt.xlabel('xxxx')
plt.ylabel('yyyy')

# 
plt.legend()
# 
plt.title('cos&sin')

plt.show()
```
