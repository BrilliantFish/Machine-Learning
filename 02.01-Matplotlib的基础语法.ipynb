```python
x = np.linspace(-10, 10, 100)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1, color = 'green') # 传入颜色
plt.plot(x,y2, linestyle = ':') # 设置为虚线

plt.show()
```

```python
plt.plot(x,y1, color = 'green') # 传入颜色
plt.plot(x,y2, linestyle = ':') # 设置为虚线

# 但希望图的范围不要太大时，我们可以设置x,y范围
plt.xlim(-5,5)
plt.ylim(-2,2)

plt.show()
```

```python
plt.plot(x,y1, color = 'green') # 传入颜色
plt.plot(x,y2, linestyle = ':') # 设置为虚线

# 但希望图的范围不要太大时，我们可以设置x,y范围
# plt.xlim(-5,5)
# plt.ylim(-2,2)
# 也可以这样
plt.axis([-5,5,-2,2])

plt.show()
```

## 给X,Y轴设置名字 & 给线设置名字

```python
plt.plot(x,y1, color = 'green', label='sin(x)') # 传入颜色
plt.plot(x,y2, linestyle = ':', label='cos(x)') # 设置为虚线

plt.xlabel('xxxx')
plt.ylabel('yyyy')

# 调用图例，显示曲线的名称
plt.legend()
# 设置图标名字
plt.title('cos&sin')

plt.show()
```

## 离散情况

```python
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.show()
```

```python
x = np.random.normal(0,100,10000)
y = np.random.normal(0,100,10000)
plt.scatter(x,y, alpha = 0.1) # 设置不透明度， 可以更加清晰的看到分布的密度，1是不透明，0是透明
plt.show()
```
