## KNN的最近邻思想
1. KNN最早提出是应用于分类，但是在后面的发展中也适用于回归和聚类
2. “近朱者赤，近墨者黑”，如：当求一个未知点的分类，是由其周围的邻居决定，当然这个周围的计算指代是特征空间的距离，这个后面会讲。

## KNN在分类和回归应用的核心思想
#### 分类:
1. 当我们求白点A的归于红的还是黑的时，我们会认为A距离红点群体近就是红点，反之就是黑点，注意我们是选择离群体近的，而不是某个点最近的，所以我们需要找多个点。
2. 但是我们可以看到，假如我们选择3个最近的点时，红色占多数，可以认为时红色，假如我们选择5个最近的点时，黑色占多数，可以认为时黑色（民主集中投票制，权重也不一定是一人一篇）
3. 所以我们可以看到K值的选择会导致不同的准确率
4. 而假如我们选择的K太小的，会造成邻居少，依赖性某一类邻居的程度过高，导致过拟合（过拟合下面会解释），而选择的K很大，就会导致红色也可以，黑色也可以，就无法做到“近朱者赤近墨者黑”
> 你可能回想为什么K不选择全部呢，然后选择平均距离，但是这样其实是不好的，假如所有都参与，那么就变成了群体中谁的数量多谁就取胜，不符合“K近邻”&“近朱者赤近墨者黑”  



#### 回归:
1. 回归是计算一个未知点的值，所以我们可以通过周围的邻居来推测位置点的值
> 举个例子，古语有云：“物以类聚，人以群分”，假如我们要推算你的资产有多少钱，那我们可以从你经常交往的朋友&同学&行业现状&同事等，毕竟 “物以类聚，人以群分” 



## 过拟合 & 欠拟合
过拟合(overfit)：就是过分依赖某一类邻居  
欠拟合(underfit)：就是对无法判断属于什么群体邻居  
最优拟合(fit)

## 距离的度量
判断在特征空间中两点之间的距离有不同的度量法则，不同的度量法则下距离的远近不一样。
> 比如，A，B两地距离很远，年轻人想着乘坐飞机去B，他觉得很近，一小时就到了，老爷爷觉得骑自行车太久了，不想去B那里，这里就是度量方式和手段的不同导致的。

## 距离的计算方式
1. 明式距离 (Minkowski Distance)
2. p=1时，曼哈顿距离(Manhatan Distance)
3. p=2时，欧式距离(Euclidean Distance)
