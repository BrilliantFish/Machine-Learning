## 多项式回归
#### 本章内容
* 介绍分析与处理多项式的手法
* sklearn 进行多项式回归编程
* 使用sklearn进行多项式的特征向量预处理

#### 分析与处理多项式的手法
* 我们看到下图是一个多项式，但是只有一个特征向量$x$，这个时候我们可以再构造一个特征向量$x2$, 原本特征向量作为$x1$，只是$x2$是$x1^2$
* 这样我们就变成了处理两个特征向量的多元线性回归了

## 代码实现

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## 构造非线性数据
x = np.random.uniform(-4, 2, size = (100))
y = 2*x**2 + 4*x + np.random.randn(100)

X = x.reshape(-1,1)
```

```python
plt.scatter(X, y)
plt.show()
```

```python
## 这里只有一个特征向量，所以我们需要构造新的特征向量
X_new = np.hstack([X, X**2])
X_new.shape, y.shape
```

```python
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
```

```python
## 训练模型
lin_reg = LinearRegression()
model = lin_reg.fit(X_new, y)
model
```

```python
y_predict = model.predict(X_new)
```

```python
plt.scatter(x,y)
plt.plot(np.sort(x), y_predict[np.argsort(x)],color='red')
plt.show()
```

### 采用sklearn实现特征向量的多项式预处理 - PolynomialFeatures

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_features = PolynomialFeatures(degree=2)  # 这里的degree表示新构建的特征向量是原特征向量的几次方，这里是三次方

## 这里你会发现是三个特征值，但是第一个特征值是1，代表x0
X_new = polynomial_features.fit_transform(X)
X_new
```
