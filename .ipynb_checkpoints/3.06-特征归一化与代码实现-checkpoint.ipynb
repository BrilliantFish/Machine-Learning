## 特征归一化的概念与作用
1. 特征归一化(Feature Scaling)，
2. 如下图所示，有两个特征值“身高”“体重”来判断这个人是【瘦，正常，胖】，如果按照下面单位的数值来计算两个点的距离就会有一个很奇怪的点，就是距离基本由“体重差”决定，“身高”在这里的决定权重太低了
3. 得出概念：特征归一化并不是让每个特征的权重都变成一样，而是让每个特征的尺度或者分布都在一个类似的范围内。

## 特征归一化方法 

#### 最大最小值归一化
1. 先找出特征集的最大值和最小值，然后每个特征值-最小值，除以最值的差，最终得到归一化
2. 采用这种方法会将特征范围变成【0，1】

* 缺点：
    1. 假如体重特征集中有一个最大值400，其他人一般是在80-150，那就会导致我们的归一化结果不会超过0.21，分布很集中，变成大家都差不多
    2. 所以说假如特征集中有一个很突出，就会导致我们的归一化之后的值变得很集中，这个特征就没有什么价值了，因为计算距离时做差基本等于零  
      
* 适用情况：
    1. 特征范围是确定的，而且彼此差距之间比较均匀，比如考试成绩最低0，最高100，最异常不会超过100，比如200之类的

#### 零均值归一化
1. 会将特征集映射为方差为0，标准差为1的分布，也就是正态分布

* 适用范围：
    1. 除非数据有明显的边界情况，否则我们一般采用这种方法，也就是说我们一般采用这种方法执行特征归一化

## 代码实现

#### 最大最小值归一化

```python
import numpy as np
from sklearn import datasets
```

```python
## 导入iris数据集
iris = datasets.load_iris()
```

```python
## 拿到样本特征&样本标记
X = iris.data
y = iris.target
X, y
```

```python
## 计算各个特征的最大值
x_max = np.max(X, axis=0)
x_max 
```

```python
## 计算各个特征的最小值
x_min = np.min(X, axis=0)
x_min
```

```python
## 最大最小值归一化
X = (X-x_min)/(x_max - x_min)
X
```

#### 零均值归一化

```python
## 导入iris数据集
iris = datasets.load_iris()
```

```python
## 拿到样本特征&样本标记
X = iris.data
y = iris.target
X, y
```

```python
# 计算均值
X_mean = np.mean(X, axis = 0)
X_mean
```

```python
## 计算标准差
X_std = np.std(X, axis=0)
X_std
```

```python
X_new = (X-X_mean)/X_std
X_new
```

```python
# 看是不是正态分布
np.mean(X_new, axis=0)
```

```python
np.std(X_new, axis=0)
```

```python
## 均值基本为0， 标准差为1，属于正态分布
```

## 采用scikit-learn中的StandardScalar实现零值归一化

```python
from sklearn.preprocessing import StandardScaler
```

```python
## 导入iris数据集
iris = datasets.load_iris()
```

```python
## 拿到样本特征&样本标记
X = iris.data
y = iris.target
X, y
```

```python
# 创建一个StandardScaler实例
standard_scaler = StandardScaler()
```

```python
# 调用fit计算均值和标准差，数据存储在standard_scaler中
standard_scaler.fit(X)
```

```python
## 均值
standard_scaler.mean_
```

```python
## 标准差
standard_scaler.scale_
```

```python
# 调用standard_scaler直接计算零值归一化数据
X_new = standard_scaler.transform(X)
X_new
```
