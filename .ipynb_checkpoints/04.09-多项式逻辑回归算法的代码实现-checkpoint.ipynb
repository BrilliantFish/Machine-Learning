#### 数据集

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
x = np.random.normal(0,1,size=(200,2))
y = np.array((x[:,0]) + (x[:,1]**2)<2, dtype = 'int')
y
```

```python
from sklearn import model_selection
## 划分数据集
x_train,x_test, y_train, y_test = model_selection.train_test_split(x,y,random_state=233,train_size=0.7)
```

```python
## 样本分布
plt.scatter(x_train[:,0], x_train[:,1], c=y_train)
plt.show()
```

#### 预处理特征数据

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
ploy = PolynomialFeatures(degree=2)
```

```python
ploy.fit(x_train)
```

```python
x2_train = ploy.transform(x_train)
x2_test = ploy.transform(x_test)
```

#### 逻辑回归

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
lg.fit(x2_train, y_train)
```

```python
lg.score(x2_train, y_train)
```

```python
lg.score(x2_test, y_test)
```
