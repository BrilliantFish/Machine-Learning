## 生成数据集

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# 生成数据集
x, y = make_blobs(
    n_samples = 300,         # 样本的总数
    n_features = 2,          # 特征的种类
    centers=3,                # 样本的种类，也就是y的种类
    cluster_std=1,           # 这个参数控制每个类别或者集群的标准偏差，会影响生成数据的分散程度，标准偏差越大，每个类别的数据越分散。
    center_box=[-10,10],     # 取值的范围，特征值取值范围
    random_state=233,        # 数据的种子，因为是随机生成的，为了可以再次创建一样的数据，在别的地方创建时，可以采用同样的配置与种子
    return_centers=False     # 暂时先不用理解，后面会用到再解释
)
```

```python
# 展示数据集分布
#c是分类的颜色，plt会对y的值进行分类， s是图上点的大小  
#我们也可以进行分类后调用三次scatter，像上一节那样，但是麻烦
plt.scatter(x[:,0], x[:,1], c=y, s=15) 

plt.show()
```

## 划分数据集

```python
# 我们需要先将数据集打乱，再从中分训练集和测试集，为了不对原数据进行打乱，我们只拿打乱后的索引数组
# 设置随机种子
np.random.seed(233)
# 打乱后的索引数组
shuffle = np.random.permutation(len(x))
shuffle
```

```python
#设置训练集比例
train_size = 0.7
```

```python
# 训练集的索引
train_index = shuffle[:int(len(shuffle)*train_size)]

# 测试集的索引
test_index = shuffle[int(len(shuffle)*train_size):]

train_index.shape, test_index.shape
```

```python
# 分割训练集
x[train_index].shape, y[train_index].shape
```

```python
# 分割测试集和
x[test_index].shape, y[test_index].shape
```

### 以上可以封装成一个函数，实现样本输入，自动返回训练集和测试集

```python
def my_train_test_split(x,y,train_size=0.7, random_state=None):
    if random_state: 
        np.random.seed(random_state)
    shuffle = np.random.permutation(len(x))
    train_index = shuffle[:int(len(shuffle)*train_size)]
    test_index = shuffle[int(len(shuffle)*train_size):]
    return x[train_index], y[train_index], x[test_index], y[test_index]
```

```python
# 调用我们自己封装的函数
x_train,y_train,x_test,y_test = my_train_test_split(x,y,random_state=233)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
# 看看训练集和测试集的分布情况
plt.scatter(x_train[:,0],x_train[:,1],c=y_train,s=15)
plt.show()
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c=y_test,s=15)
plt.show()
```

## 使用sklearn划分数据集和测试集

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,random_state=233,train_size=0.7)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# 看看训练集和测试集的分布情况
plt.scatter(x_train[:,0],x_train[:,1],c=y_train,s=15)
plt.show()
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c=y_test,s=15)
plt.show()
```
