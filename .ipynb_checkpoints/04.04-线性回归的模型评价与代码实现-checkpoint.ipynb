#### RMSE
RMSE就是对MSE进行开方，实现量纲一致，比如MSE的话就是平方米，RMSE就是米

#### 代码实现

```python
### 导入数据
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets 
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
```

```python
## 获取Boston数据集, sklearn 1.2之后，不能像获取鸢尾花数据一样load_boston(), 这个函数被废除了，可以采用这种方式获取数据
## 要安装pandas包
boston = fetch_openml(name="boston", version=1, as_frame=True, parser='auto')
## 拿数据
x = boston.data.values
y = boston.target
x = x[y<50]
y = y[y<50]


## 划分数据集
x_train,x_test, y_train, y_test = model_selection.train_test_split(x,y,random_state=233,train_size=0.7)

## 采用sklearn的线性回归库
lin_reg = LinearRegression()
model = lin_reg.fit(x_train, y_train)

## 获取预测值
y_predict = model.predict(x_test)
```

#### MSE

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_predict)
```

#### RMSE

```python
mean_squared_error(y_test, y_predict, squared=False)
```

#### MAE

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, y_predict)
```

$R^2$

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_predict)
```
