## 逻辑回归（多元逻辑回归 & 多项式逻辑回归）
找出不同种类样本之间的决策边界（分割线）

#### 注意
1. 在学习逻辑回归的时候我们应该清晰的知道，我们是要利用线性算法来进行分类，所以才叫逻辑回归，这时候我们图例上的坐标就不在是y值和x值了，在分类问题中，坐标轴都是用来展现特征向量的，而线性回归中才会有一个轴表示y值。
2. 线性回归中，会直接求回归线，拟合各个特征，但是在逻辑回归中，我们并不会直接求回归线（决策边界）,我们会先设定各个特征的相关性，在下图中，这个t就是各个特征向量的相关性函数：$t = f(x)$
3. 这个 $Sigmod$ 函数会讲我们的相关性t转化成 $[0, 1]$ 区间的值，所以我们的就是通过将训练集的数据分成 $[0, 0.5], [0.5, 1]$ 之间来表现分类，调整$W$（特征的权重向量），使得我们的训练集数据按照预期分类，最终就会得到最终模型的特征权重向量$W$
4. 最终$W$就会得到最终的相关性函数：$t = f(x)$，而数学家发现，这个相关性函数就是决策边界！！！

#### 损失函数
* 当样本应为$y=1$时，也就是说p应该越趋近于1
* 当样本应为$y=0$时，也就是说p应该越趋近于0
