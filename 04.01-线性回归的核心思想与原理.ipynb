## 线性算法的核心思想
1. 就是找出各个样本之间的关系，这个关系在图像上可能是线性关系，可能是曲线关系，假如我们能找到这样的关系，那我们就能找到他们的规律，也就是说我们给出一组特征值，就能得到预测值（这个就叫回归，回归的意思就是找到了样本的规律和本质，回归原型）
2. 也就是说线性算法都是用来处理回归问题的，也就是说用来预测的（线性回归）
3. 但是可以使用线性算法来处理分类问题（逻辑回归），它的原理是通过值，给每个区间的值设定一个分类标记A、B、C、D，也就是说当预测值落在某个区间的时候，我们就可以得到它的分类，可以看到这种手法也是采用线性算法实现的

## 线性回归（一元线性回归 & 多元线性回归 & 多项式逻辑回归）
找出样本之间的关系，形成拟合线，进行预测值

#### 一元线性回归
* 你会发现我们在找各个样本点到拟合线的最近距离时，不是点到直线的垂直距离，这是因为我们这里是$\hat{y}$ (真实值)与$y$ (预测值)之间的差距，同一个$x$下的才好进行比较，而且注意这里是预测值，所以采用同一个$x$下的值，所以误差：$|y-\hat{y}|$
* 我们会通过最小二乘法计算出最优参数，也就是最优化
* 为什么这里的最优化会叫凸优化呢？这是因为这里的损失函数是一个凸函数，也就是说有唯一最小值，而且可以通过求导求0的方式找到最小点（这种通过求导求0方式并可以通过加减乘除等基本运算得到最值叫做闭式解），这种方式就叫凸优化，但是有些损失函数虽然是一个凸函数，也就是说有唯一最小值，不存在局部最小值，虽然可以求导，但是却无法执行闭式解，主要原因是有一些奇奇怪怪的公式（log和逻辑函数等），所以这种的求解就无法很简单的计算出来，得用别的方式，比如逻辑回归的损失函数，后面会讲
